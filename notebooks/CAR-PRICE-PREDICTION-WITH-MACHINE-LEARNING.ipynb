# ExtraTreesRegressor

```python
#ExtraTreesRegressor
from sklearn.ensemble import ExtraTreesRegressor
extra_reg= ExtraTreesRegressor()
extra_reg.fit(X_train,y_train)
y_pred = extra_reg.predict(X_test)
print("Accuracy on Training set", extra_reg.score(X_train,y_train))
print("Accuracy on Testing set", extra_reg.score(X_test, y_test))
```

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error

print("\t\tError Table")
print('Mean Absolute Errsor      : ', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared  Error      : ', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared  Error : ', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R Squared Error          : ', metrics.r2_score(y_test, y_pred))
```
